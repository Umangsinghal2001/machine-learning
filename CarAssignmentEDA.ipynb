# Preprocessing

```python
df['doornumber'] = df['doornumber'].map({'two': 2, 'four': 4})
df['cylindernumber'] = df['cylindernumber'].map({'two': 2, 'three': 3, 'four': 4, 'five': 5, 'six': 6, 'eight': 8, 'twelve': 12})
```

```python
# creating features and label variable

X = df.drop(columns = 'price', axis = 1)
y = df['price']
```

```python
X = pd.get_dummies(X, drop_first = True)
X.head()
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test) 
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
lr.score(X_train, y_train)
```

```python
lr.score(X_test, y_test)
```

```python
from sklearn.linear_model import Lasso, LassoCV
```

```python
lassocv = LassoCV(alphas = None, cv = 10, max_iter = 10000, normalize = True)
lassocv.fit(X_train, y_train)
```

```python
lasso = Lasso(alpha = lassocv.alpha_)
lasso.fit(X_train, y_train)
```

```python
lasso.score(X_train, y_train)
```

```python
lasso.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor

rfr = RandomForestRegressor()
rfr.fit(X_train, y_train)
```

```python
rfr.score(X_train, y_train)
```

```python
rfr.score(X_test, y_test)
```

### Conclusion : Random forest has achieved highest score of 90% on test data
