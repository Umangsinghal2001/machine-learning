```python
import pandas as pd
import matplotlib.pyplot as pit
```

```python
df=pd.read_csv('archive1/ml-100k/u.data',sep='\t',names=['user_id','item_id','rating','ts'])
```

```python
df.head()
```

```python
cols="item_id|title| release date | video release date |\
              IMDb URL | unknown | Action | Adventure | Animation |\
              Children's | Comedy | Crime | Documentary | Drama | Fantasy |\
              Film-Noir | Horror | Musical | Mystery | Romance | Sci-Fi |\
              Thriller | War | Western |".split(sep="|")
movies=pd.read_csv("archive1/ml-100k/u.item",sep="|",names=cols)
```

```python
movies.head()
```

```python
df2=df.drop("ts",axis=1)
data=df2.merge(movies[["title",'item_id']],on='item_id')
```

```python
data.head()
```

```python
data.describe()

```

```python
data
```

```python
df_pivot= data.pivot_table(index="user_id",columns="title",values='rating') 
```

```python
df_pivot.head()
```

```python
inp="Star Wars (1977)"
```

```python
df_pivot[inp]
```

```python
cor=df_pivot.corrwith(df_pivot[inp])
```

```python
cor
```

```python
sim_df=pd.DataFrame(cor,columns=["correlations"])
```

```python
sim_df.head()
```

```python
sim_df.sort_values(by="correlations",ascending=False).head(20)
```

```python
avg_rates=data.groupby('title')["rating"].mean()
```

```python
avg_rates
```

```python
rate_count=data.groupby("title")["rating"].count()
```

```python
rate_count
```

```python
sim_df["rate_count"]=rate_count
```

```python
sim_df["avg_rates"]=avg_rates
```

```python
sim_df
```

```python
recomm=sim_df[sim_df["rate_count"]>100].sort_values(by="correlations",ascending=False)
```

```python
recomm.drop(inp,axis=0,inplace=True)
```
